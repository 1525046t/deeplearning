```python
import numpy as np
```

```python
'''単純パーセプトロン''' 

class SimplePerceptron(object):
    
    # 初期化(次元,重み,バイアス)
    def __init__(self,input_dim):
        self.input_dim=input_dim
        self.w=np.random.normal(size=(input_dim,))
        self.b=0.
        
    # 前進計算
    def forward(self,x):
        y=step(np.matmul(self.w,x)+self.b)
        return y

    # 誤差計算
    def compute_deltas(self,x,t):
        y=self.forward(x)
        delta=y-t
        dw=delta*x
        db=delta
        return dw,db
    
# ステップ関数
def step(x):
    return 1*(x>0)

# メイン関数
if __name__=='__main__':
    np.random.seed(123)
    
    '''
    1.データの準備
    '''
    # 入力次元
    d=2
    # 教師データ数
    N=20
    
    # 入力データ
    x1=np.random.randn(N//2,d)+np.array([0,0])
    mean=5
    x2=np.random.randn(N//2,d)+np.array([mean,mean])
    # 教師データ
    t1=np.zeros(N//2)
    t2=np.ones(N//2)
    
    # 結合
    x=np.concatenate((x1,x2),axis=0)
    t=np.concatenate((t1,t2))
    
    '''
    2.モデルの構築
    '''
    
    model=SimplePerceptron(input_dim=d)
    
    '''
    3.モデルの学習
    '''
    
    # 損失判定
    def compute_loss(dw, db):
        return all(dw == 0) * (db == 0)
    
    # 誤差逆伝搬
    def train_step(x, t):
        dw, db = model.compute_deltas(x, t)
        loss = compute_loss(dw, db)
        model.w = model.w - dw
        model.b = model.b - db
        return loss

    # 反復
    while True:
        classified = True
        for i in range(N):
            loss = train_step(x[i], t[i])
            classified *= loss
        if classified:
            break

    '''
    4. モデルの評価
    '''
    
    print('w:', model.w) 
    print('b:', model.b) 

    print('(0, 0) =>', model.forward([0, 0])) 
    print('(5, 5) =>', model.forward([5, 5]))
```
